```python
import pandas as pd

import numpy as np

from sklearn.model_selection import train_test_split
```

```python
df_acute = pd.read_csv('../datasets/acute/diagnosis.csv',
                       engine='c', 
                       memory_map=True, 
                       low_memory=True)
```

```python
X = df_acute[['temperatura', 'nausea', 'dorlombar', 
              'urinepushing', 'miccao', 'queimacao', 'inflamacao']]
# Labels
y = df_acute['target']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, 
                                                    random_state=100, 
                                                    shuffle=True, 
                                                    stratify=y)
```

```python
# testa com enginnering
model = BaseEnginnering()
model.train_X = X_train
model.train_y = y_train
print('Shape X: ', model.train_X.shape)
print('Shape y:', model.train_y.shape)
model.train_X.head()
```

```python
# habilita a codificação dos dados (X)
model.encoder_enable = True
model.encoder_data = True
print('Shape Before: ', model.train_X.shape)
model.get_transform()
print('Shape After: ', model.train_X.shape)
print(model.train_X)
```

```python
# habilita a codificação do alvo (y)
model.encoder_enable = True
model.encoder_target = True
print('Shape y Before: ', model.train_y.shape)
print(model.train_y)
print(type(model.train_y))
model.get_transform()
print('Shape y After: ', model.train_y.shape)
print(model.train_y)
```
